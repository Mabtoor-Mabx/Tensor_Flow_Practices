```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model
```

```python
X = tf.range(-100,100,4)
X
```

```python
Y= X+10
```

```python
Y
```

```python
len(X)
```

```python
len(Y)
```

```python
X_train = X[:40]
Y_train= Y[:40]
X_test= X[40:]
Y_test  = Y[40:]


```

```python
plt.figure(figsize=(10,7))
plt.scatter(X_train, Y_train, color='red', label='The Train Values')
plt.scatter(X_test, Y_test, color='blue', label='The test Values')
plt.legend()
plt.show()
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(1, input_shape=[1])
])
model.compile(
    loss= tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(),
    metrics=['mae']
)
```

```python
model.summary()
```

```python
model.fit(tf.expand_dims(X, axis=1), Y, epochs=100 )
```

```python
model.summary()
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred.shape
```

```python
y_pred_squeeze=  tf.squeeze(y_pred)
```

```python
y_pred_squeeze.shape
```

```python
X_test.shape
```

```python
def plot_prediction(
    train_data = X_train,
    train_label= Y_train,
    test_data = X_test,
    test_label = Y_test,
    predictions=  y_pred_squeeze):
  plt.scatter(train_data, train_label, color='red', label='The Train values')
  plt.scatter(test_data, test_label, color='green', label='The Test Values ')
  plt.scatter(predictions, test_data, color='blue', label='Predicted value')
  plt.legend()
  plt.title("Model Of Plot Predictions")
  plt.xlabel('X values')
  plt.ylabel('Y values')
  plt.show()

plot_prediction();
```

# **Model No 1**

```python
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
                             tf.keras.layers.Dense(1)
])

model_1.compile(
    loss= tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(),
    metrics=['mae']
)

model_1.fit(tf.expand_dims(X, axis=1), Y, epochs=100)
model_1.evaluate(X,Y)
```

```python
y_pred_1 = model_1.predict(X_test)

plot_prediction(predictions=y_pred_1)
```

```python
def mae(y_test, y_pred):
  return tf.metrics.mean_absolute_error(y_test=Y_test, y_pred=y_pred)
```

```python
def mse(y_test, y_pred):
  return tf.metrics.mean_squared_error(Y_test = Y_test, y_pred = y_pred)
```
