```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model
```

# Neural Network Classification

```python
from sklearn.datasets import make_circles
```

```python
n_samples = 1000
```

```python
X,Y =   make_circles(
    n_samples,
    noise=0.02,
    random_state=42
)
```

```python
X
```

```python
y[:10]
```

# Create Dataset

```python
dataset = pd.DataFrame({"X0": X[:,0],"X1": X[:,1],"Label": Y})
```

```python
dataset
```

# Plot A Graph

```python
# plt.scatter(X[:,0], X[:,1], color='y', cmap=plt.cm.RdYlBu)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu);
```

# Build Neural Classification Network

```python
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
                               tf.keras.layers.Dense(1)
])

model_1.compile(
    loss = tf.keras.losses.BinaryCrossentropy(), 
    optimizer = tf.keras.optimizers.SGD(),
    metrics=['accuracy']
)

model_1.fit(tf.expand_dims(X,axis=1), Y, epochs=5)
```

```python
model_1.evaluate(X,Y)
```

# Model 2

```python
tf.random.set_seed(42)

model_2 = tf.keras.Sequential([
                               tf.keras.layers.Dense(1),
                               tf.keras.layers.Dense(1)
])

model_2.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.SGD(),
    metrics=['accuracy']
)

model_2.fit(tf.expand_dims(X,axis=1), Y, epochs=200)
```

```python
model_2.evaluate(X,Y)
```

```python
model_1.evaluate(X,Y)
```

# Model No 3

```python
tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
                               tf.keras.layers.Dense(100, activation='relu'),
                               tf.keras.layers.Dense(10),
                               tf.keras.layers.Dense(1), 
])


model_3.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

model_3.fit(tf.expand_dims(X,axis=1),Y, epochs=500)
```

```python
model_1.evaluate(X,Y)
```

```python
model_2.evaluate(X,Y)
```
