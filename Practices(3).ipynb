##Random

```python
random = tf.random.Generator.from_seed(42)
random = random.normal(shape=(1,2))

random
```

```python
random2 = tf.random.Generator.from_seed(42)
random2 = random2.uniform(shape=(1,2))

random2
```

```python
random==random2
```

#Shuffle
