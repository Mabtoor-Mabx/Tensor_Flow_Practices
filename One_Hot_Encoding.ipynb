```python
import tensorflow as tf
import numpy as np
```

```python
listes= [1,2,3,4,5,6,7]
tf.one_hot(listes, depth=8)
```

```python
iset =  [2,3,5,7,8,3,1]
one_hot_values = tf.one_hot(iset, depth=6, on_value=1234, off_value=2233)
one_hot_values
```

```python
liss  = [2,3,4,5,6,8,9]
tf.one_hot(liss, depth=5, off_value=111, on_value=000, dtype=tf.int32)
```
