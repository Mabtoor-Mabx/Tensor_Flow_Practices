**Random Generator**

```python
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=(4,5))
random_1
```

**Shuffle**

```python
shuffling = tf.constant([
                         [3,4],[5,3],[6,3]
])
not_shuffled = tf.random.shuffle(shuffling)
not_shuffled
```

```python
not_shuffled = tf.random.shuffle(shuffling, seed=42)
not_shuffled
```

```python
tf.random.set_seed(42)
not_shuffled = tf.random.shuffle(shuffling, seed=42)
not_shuffled
```

**Matrix Multiplication**

```python
matrix1 = tf.constant([
                       [2,4],[6,4],[6,3]
])
```

```python
matrix2 = tf.constant([
                       [3,4],[7,4],[8,2]
])
```

```python
tf.matmul(matrix1, matrix2)
```

```python
tf.matmul(tf.transpose(matrix1), matrix2)
```

```python
tf.matmul(matrix1, tf.transpose(matrix2))
```

```python
matrix1.shape, matrix2.shape
```

```python
tf.reshape(matrix1, shape=(2,3))
```

```python
tf.matmul(tf.reshape(matrix1, shape=(2,3)), matrix2)
```

```python
tf.matmul(matrix1, tf.reshape(matrix1, shape=(2,3)))
```

**Aggregate**

```python
numpy_B = tf.constant([
                       [-1,-4],[-7,-5],[-7,-3]
])
numpy_B
```

```python
tf.abs(numpy_B)
```

```python
print(tf.reduce_min(numpy_B))
print(tf.reduce_max(numpy_B))
print(tf.reduce_mean(numpy_B))
print(tf.reduce_sum(numpy_B))
```

**Variance and Standard Deviation**

```python
tf.math.reduce_variance(tf.cast(numpy_B, dtype=tf.float32))
```

```python
tf.math.reduce_std(tf.cast(numpy_B, dtype= tf.float32))
```
