```python
import tensorflow as tf

not_shuffled = tf.constant([
  [1,2], [3,5], [6,7], [8,9],
])

not_shuffled
```

```python
tf.random.shuffle(not_shuffled)
```

```python
tf.random.shuffle(not_shuffled, seed=42)
```

```python
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled , seed=42)
```

```python
shuffle1= tf.constant([
                       [1,3],[4,5],[6,8],[1,4]
])

shuffle2= tf.constant([
                       [3,1],[4,6],[2,5],[6,9]
])

shuffle3= tf.constant([
                       [3,6],[1,8],[3,7],[2,9]
])

shuffle4= tf.constant([
                       [3,1],[4,6],[2,5],[6,9]
])

shuffle5= tf.constant([
                       [4,1],[2,7],[1,9],[4,7]
])


tf.random.set_seed(42)

tf.random.shuffle(shuffle1 ,seed=42)
tf.random.shuffle(shuffle2 ,seed=42)
tf.random.shuffle(shuffle3, seed=42)
tf.random.shuffle(shuffle4 ,seed=42)
tf.random.shuffle(shuffle5 ,seed=42)

shuffle1, shuffle2, shuffle3, shuffle4, shuffle5
```
