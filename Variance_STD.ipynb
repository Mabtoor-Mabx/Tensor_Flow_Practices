#**Variance and Standard Deviation**

```python
import tensorflow as tf
import numpy as np
import tensorflow_probability as tfp
```

```python
matrix_1 = tf.constant([
                        [1,3],[5,3],[8,7]
])
```

```python
matrix_2 = tf.constant([
                        [1,4],[6,3],[7,4]
])
```

```python
matrix_1, matrix_2
```

```python
tfp.stats.variance(matrix_1)
```

```python
tfp.stats.variance(matrix_2)
```

```python
# tf.math.reduce_std(matrix_1)
```

```python
tf.math.reduce_std(tf.cast(matrix_1, dtype=tf.float32))
```

```python
tf.math.reduce_std(tf.cast(matrix_2, dtype=tf.float16))
```
