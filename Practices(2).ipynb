#Random

```python
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=(2,3))
random_1
```

```python
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(2,4))
random_2
```

```python
random_1==random_2
```

```python
random_3 = tf.random.Generator.from_seed(42)
random_3 = random_3.uniform(shape=(2,4))
random_3
```

```python
random_2==random_3
```
