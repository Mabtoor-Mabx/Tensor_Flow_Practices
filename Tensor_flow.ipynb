**Matrix Multiplications**

```python
matrix_1 = tf.constant([
                        [4,4],[6,3],[2,3]
])
matrix_1
```

```python
matrix_2 = tf.constant([
                        [4,6],[8,5],[3,2]
])
matrix_2
```

```python
tf.matmul(matrix_1,matrix_2)
```

```python
tf.matmul(tf.transpose(matrix_1), matrix_2)
```

```python
tf.matmul(matrix_1, tf.transpose(matrix_2))
```

```python
matrix_1.shape
```

```python
tf.transpose(matrix_2)
```

**Aggregate**

```python
tf.abs(matrix_2)
```

```python
print(tf.math.reduce_min(matrix_1))
print(tf.math.reduce_mean(matrix_1))
print(tf.math.reduce_max(matrix_1))
print(tf.math.reduce_sum(matrix_1))
```

**Varianve and Std**

```python
tf.math.reduce_variance(tf.cast(matrix_1, dtype=tf.float32))
```

```python
tf.math.reduce_std(tf.cast(matrix_1, dtype=tf.float16))
```
