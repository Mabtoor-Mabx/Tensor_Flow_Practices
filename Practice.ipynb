# **Squueze, One_Hot, More_Math & Tensor_To_Numpy and Vice_Versa**

```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
squeezed = tf.constant(tf.random.uniform(shape=[50]), shape=(1,1,1,1,50))
```

```python
squeezed
```

```python
squeezed_value = tf.squeeze(squeezed)
```

```python
squeezed_value
```

```python
val2 = tf.constant(tf.random.uniform(shape=[20]), shape=(1,1,1,1,20))
val2
```

```python
val2_vale = tf.squeeze(val2)
val2_vale
```

```python
vall = tf.constant([
                     [2,4],[5,3],[5,7],[8,6]
])
vall
```

```python
tf.one_hot(vall, depth=5)
```

```python
tf.one_hot(vall, depth=4, on_value = 'I am On value', off_value='I am Off value')
```

```python
tf.math.log(tf.cast(vall, dtype=tf.float32))
```

```python
tf.math.sin(tf.cast(vall, dtype=tf.float32))
```

```python
tf.math.cos(tf.cast(vall, dtype=tf.float32))
```
