```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.utils  import plot_model
```

# Model No 1

```python
X = tf.constant([2,4,6,8,10,12,14])
Y= tf.constant([3,6,9,12,15,18,21])
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(1)
])

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(),
    metrics=['mae']
)

model.fit(tf.expand_dims(X,axis=1), Y, epochs=5)

model.evaluate(X,Y)
```

```python
y_pred_squeeze= tf.squeeze(y_pred)
y_pred_squeeze
```

```python
def plot_prediction(
    train_data = X_train,
    train_label = Y_train,
    test_data = X_test,
    test_label = Y_test,
    predictions= y_pred_squeeze):
  
  plt.scatter(train_data, train_label, color='red', label='Train Values' )
  plt.scatter(test_data, test_label , color='green', label='Test Values')
  plt.scatter(test_data,predictions, color='blue', label='Test Values With Predictions')
  plt.legend()
  plt.title('Train and Test Values')
  plt.xlabel('Train Values')
  plt.ylabel('Test Values')
  plt.show()


plot_prediction();
```
