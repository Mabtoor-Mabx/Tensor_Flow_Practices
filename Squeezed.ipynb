```python
import tensorflow as tf
import numpy as np
```

```python
G = tf.constant(tf.random.uniform(shape=[50]), shape=(1,1,1,1,50))
G
```

```python
tf.random.set_seed(42)
G = tf.constant(tf.random.uniform(shape=[50]), shape=(1,1,1,1,50))
```

```python
G
```

```python
G_squeezed = tf.squeeze(G)
G_squeezed
```
