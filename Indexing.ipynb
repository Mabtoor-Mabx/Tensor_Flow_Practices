##Indexing and NewAxis In Tensors

```python
import tensorflow as tf
import numpy as np
```

```python
some_list = [1,2,3,4,5,6,7,8,9,0]
some_list
```

```python
some_list[:1]
```

```python
some_list[:0]
```

```python
some_list[:2]
```

```python
rank_4_tensor = tf.zeros(shape=(2,3,4,6))

rank_4_tensor
```

```python
rank_4_tensor[:1,:1,:1,:]
```

```python
rank_4_tensor[:1,:,:,:1]
```

```python
rank_4_tensor[...]
```

```python
rank_3_tensor = rank_4_tensor[...,tf.newaxis]
```

```python
rank_3_tensor
```

```python
rank_3_tensor==rank_4_tensor
```

```python
tf.expand_dims(rank_3_tensor,axis=1)
```
